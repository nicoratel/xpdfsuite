## Extraction PDF

exemples de syntaxe utilsant les fonctions décrites dans xpdfsuite.xpdfpipeline

```python
from xpdfsuite.xpdf_pipeline import extract_xpdf_from_h5, extract_xpdf_from_image
from xpdfsuite.filereader import load_h5_data
from matplotlib import pyplot as plt

ponifile = os.path.join(datapath, 'mycalib.poni')

ref_array=None

# Extract pdf from h5: averages the stack of images and performs background subtraction using a reference file or numpy array.


# Référence fournie comme fichier HDF5
r, G = extract_xpdf_from_h5(file, ponifile, 'CeO2')
                             # ref='background.h5')
plt.figure()
plt.plot(r, G)
# Référence fournie comme tableau numpy
r, G = extract_xpdf_from_h5(file, ponifile, 'CeO2',
                              ref=ref_array)
plt.figure()
plt.plot(r, G)
# extract_xpdf_from_image
info, image_array = load_h5_data(file) 

r, G = extract_xpdf_from_image(image_array, ponifile, 'CeO2')
                                #ref='background.h5')
plt.figure()
plt.plot(r, G)
r, G = extract_xpdf_from_image(image_array, ponifile, 'CeO2',
                                ref=ref_array)
plt.figure()
plt.plot(r, G)
    
```
